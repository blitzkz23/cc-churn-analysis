## Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from helper.plot_functions import *
```

## Import Data

```python
df_status = pd.read_csv('../../data/customer_distribution/status_dist.csv', delimiter='\t')
df_card = pd.read_csv('../../data/customer_distribution/card_dist.csv', delimiter='\t')
df_age = pd.read_csv('../../data/customer_distribution/age_group_dist.csv', delimiter='\t')
df_education = pd.read_csv('../../data/customer_distribution/education_dist.csv', delimiter='\t')
df_income = pd.read_csv('../../data/customer_distribution/income_dist.csv', delimiter='\t')
df_marital = pd.read_csv('../../data/customer_distribution/marital_dist.csv', delimiter='\t')
```

## Data Visualization

### Customer Status

```python
help(plot_pie_chart)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# group the data by 'CustomerStatus' and sum the 'CustomerCount' for each group
status_counts = df_status.groupby('CustomerStatus')['CustomerCount'].sum()

plot_pie_chart(status_counts, status_counts.index, 'CustomerStatus')
```
